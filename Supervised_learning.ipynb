```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(cancer.data,cancer.target,random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 3)
```

```python
cancer_fit = knn.fit(X_train,y_train)
```

```python
y_hat = cancer_fit.predict(X_test)
```

```python
print(np.mean(y_hat == y_test))
cancer_fit.score(X_test,y_test)
```

```python
from matplotlib import pyplot as plt
from pandas.plotting import scatter_matrix
scatter_matrix(df.iloc[:,:5],alpha=0.5,hist_kwds={'bins':20})
```

```python
fig,axes = plt.subplots(1,3,figsize = (10,3))
for n,a in zip((1,3,5),axes):
  clf = KNeighborsClassifier(n_neighbors=n).fit(X_train[:,:2],y_train)
  mglearn.plots.plot_2d_separator(clf,X_train,fill=True,ax = a,alpha=0.4)
  mglearn.discrete_scatter(X_train[:,0],X_train[:,:2],y_train,ax = a)
  a.set_title('{} neighbors'.format(n))
  a.set_xlabel('feature0')
  a.set_ylabel('feature1')
```
