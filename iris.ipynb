```python
from sklearn.datasets import load_iris
iris = load_iris()  #iris is not simply a dataset, it is a bunch (something similar to dictionaries)
```

```python
print('Keys: {}'.format(iris.keys()))  #keys is basically the different pieces of information in the dataset that had been loaded
```

```python
print(type(iris['data']))
print(iris['data'])
```

```python
print(iris['DESCR']+'\n..')   #Description of the dataset we are working with.
```

```python
iris.data[:6] #These are first 6 columns of the dataset
```

```python
#Split the data into train -  test sets using the
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(iris['data'], iris['target'],random_state=0,stratify=iris['target'])
# We cannot take the last 25% of the observations as the testing set as this might contain nothing but simply one class of flowers. Hence use a function that shuffles and then splits
# Random state is basically seed i.e., for reproducibity
```

```python
print(train_X.shape,test_X.shape)
print(train_y.shape,test_y.shape)
```

```python
import pandas as pd
iris_df = pd.DataFrame(train_X,columns = iris.feature_names) #columns names as feature names of the iris dataset
```

```python
from pandas.plotting import scatter_matrix
import mglearn as mg
grr = scatter_matrix(iris_df,c = train_y,figsize = (15,15),marker = 'o',hist_kwds={'bins':20}, s=60,alpha = 0.8,cmap = mg.cm3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1) #Model Defined
knn.fit(train_X,train_y) #Model fit
```

```python
knn.predict(test_X) == test_y
```

```python
import numpy as np
print("Formula applied :",np.mean(knn.predict(test_X)==test_y))
print('Using inbuilt function',knn.score(test_X,test_y))
```
